```python
import pandas as pd
import numpy as np
```

```python
data_set = pd.read_csv('model_data.csv')
```

```python
data_set = data_set.drop(data_set.columns[0], axis = 1)
```

```python
data_set
```

```python
data_set.columns[14]
```

```python
X = data_set.drop("self_shortlisted",1)
Y = data_set.pop("self_shortlisted")
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
X_final = pd.get_dummies(data= X, columns=['mhtp_x', 'consumer_gender', 'psych_gender','sliding_scale','is_visible',
                                       'bulk_billing', 'online_sessions', 'proxy_id', 'afford', 'appt_rating',
                                      'before', 'besthelp','coach', 'empathy', 'extras', 'fees100', 'fees50',
                                      'female','friend','male','mhtp_y','psygen','pubtrans', 'ready', 'sameage',
                                      'telehealth', 'workhrs'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_final, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
```

```python
model = RandomForestRegressor(n_estimators = 100,
                             max_features = "auto",
                             min_samples_leaf =3,
                             max_depth = 200,
                             random_state=42,
                             n_jobs=1,
                             oob_score=True)
```

```python
model.fit(x_train, y_train)
    
```

```python
preds = model.predict(x_test)
```

```python
def exp_rms(y_pred, target):
    target = np.exp(y_test)
    percentage_var = (target - np.exp(y_test))
    return np.mean(np.sqrt(percentage_var**2))

print("train accuracy = ", model.score(x_train, y_train))
print("val accuracy = ", model.score(x_test,y_test))
print("oob score = ", model.oob_score_)
print("rmse error = ", exp_rms(preds,y_test))
```
